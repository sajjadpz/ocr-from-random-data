```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, BatchNormalization, Activation
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.constraints import maxnorm
import pandas as pd
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from sklearn.model_selection import train_test_split
from numpy import array
import numpy as np
```

```python
#read data from csv
data = pd.read_csv("/content/drive/My Drive/tessdata.csv", names=["image", "text","confidence"])

#images i.e X
X = data['image']
#target variable
y = data['confidence']

y = y.replace(0, 1)

data.head(6)
```

```python
y
```

```python
def create_data():
    X_data, Y_data =  list(), list()
    count = 5000  
    for i, j in data.head(count).iterrows():
        image_name, confidence = X[i], y[i]
        image_id = image_name.split('.')[0]
        print("[INFO] loading images...", i, '/', count )
        filename = '/content/drive/My Drive/jpegs/out' + '/' + image_name
        image = load_img(filename, target_size=(50,50))
        image = img_to_array(image)
        X_data.append(image)
        Y_data.append(confidence)
    X_data, Y_data = array(X_data), array(Y_data)
    return X_data/255.0, Y_data
```

```python
X, Y = create_data()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
maxConfidence = y_train.max()
y_train = y_train / maxConfidence
y_test = y_test / maxConfidence
```

```python
### Creating Model
def create_model():
  model = Sequential()

  model.add(Conv2D(32, (3, 3), input_shape=(50, 50, 3), activation='relu', padding='same'))
  model.add(Activation('relu'))
  model.add(Dropout(0.2))
  model.add(BatchNormalization())

  #2nd convolutional layer
  model.add(Conv2D(64, (3, 3), padding='same'))
  model.add(Activation('relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(0.2))
  model.add(BatchNormalization())

  #3rd convolutional layer
  model.add(Conv2D(128, (3, 3), padding='same'))
  model.add(Activation('relu'))
  model.add(Dropout(0.2))
  model.add(BatchNormalization())

  model.add(Flatten())
  model.add(Dropout(0.2))

  model.add(Dense(256, kernel_constraint=maxnorm(3)))
  model.add(Activation('relu'))
  model.add(Dropout(0.2))
  model.add(BatchNormalization())
      
  model.add(Dense(128, kernel_constraint=maxnorm(3)))
  model.add(Activation('relu'))
  model.add(Dropout(0.2))
  model.add(BatchNormalization())

  #final layer should have as many neurons as there are classes. Since we have one class and it is regression problem so the value of final neuron will be between 0-99
  model.add(Dense(1))
  #model.add(Activation('softmax')) No activation function for final layer in case of regression since we need to predict values with any transformation

  optimizer = 'adam'

  model.compile(loss='mean_absolute_percentage_error', optimizer=optimizer)
  return model
```

```python
modelCNN = create_model()
```

```python
print(modelCNN.summary())
```

```python
modelCNN.fit(X_train, y_train, validation_data=(X_test, y_test),epochs=200, batch_size=8)
```

```python
print("[INFO] predicting image confidences...")
preds = modelCNN.predict(X_test)
```

```python
diff = preds.flatten() - y_test
percentDiff = (diff / y_test) * 100
absPercentDiff = np.abs(percentDiff)
```

```python
mean = np.mean(absPercentDiff)
std = np.std(absPercentDiff)
```

```python
print("[INFO] avg. confidence: {}, std confidence: {}".format(
	Y.mean(),
	Y.std()))
print("[INFO] mean: {:.2f}%, std: {:.2f}%".format(mean, std))
```
